```python
# Question 1: Dictionary keys must be immutable
my_dict = {1: "Apple", "name": "Neha", (2,3): "Tuple key"}
print("Dictionary with immutable keys:", my_dict)

```

Dictionary keys must be immutable, like numbers, strings, and tuples.
Mutable objects like lists or sets cannot be used as keys because if a key changes, the dictionary will not be able to locate its value.

```python
# Question 2: Frequency of elements in a list
elements = [1, 2, 2, 3, 1, 2]  # Input list

freq_dict = {}  

for item in elements:
    if item in freq_dict:      
        freq_dict[item] += 1   
    else:
        freq_dict[item] = 1    

print("Frequency of each element:", freq_dict)
```

We looped through each element in the list:
- >..//If the element is already in the dictionary → increase its count by 1
- >..//If the element is not in the dictionary → set its count to 1
This way, we created a frequency dictionary of the elements.

#question3:
dict[key] → Accesses the value for the key.
Problem: If the key does not exist, Python gives a KeyError.

dict.get(key) → Accesses the value for the key.
Safe: If the key does not exist, it returns None (or a value you specify).
You can also provide a default value like dict.get(key, default)

```python
# Question 3: Example showing difference

my_dict = {"a": 10, "b": 20}

# Unsafe method - this will give error if key doesn't exist
# print(my_dict["c"])  # KeyError, uncomment to test

# Safe method using get()
print("Using get() without default:", my_dict.get("c"))      
print("Using get() with default value:", my_dict.get("c", 0)) 

# Accessing existing key
print("Value of key 'a':", my_dict["a"])                     
```

- dict[key] raises an error if the key is not present in the dictionary.
- dict.get(key) returns None if the key is not found, which is safe.
- dict.get(key, default_value) allows specifying a default value if the key is missing.

```python

# Question 4: Merge two dictionaries and keep larger value for common keys

dict1 = {"a": 10, "b": 5}
dict2 = {"b": 20, "c": 30}

merged_dict = dict1.copy()

for key, value in dict2.items():
    if key in merged_dict:
        
        merged_dict[key] = max(merged_dict[key], value)
    else:
        
        merged_dict[key] = value

print("Merged dictionary with larger values for common keys:", merged_dict)
```

- First, we copied the first dictionary to a new dictionary called merged_dict.
- Then, we looped through the second dictionary.
- If a key exists in both dictionaries, we kept the larger value using max().
- If a key exists only in the second dictionary, we added it directly.
- This way, we merged both dictionaries and resolved common keys by keeping the larger value.

```python
# Question 5: Dictionary comprehension for cubes of odd numbers

numbers = [1, 2, 3, 4, 5]

# Dictionary comprehension
cube_dict = {x: x**3 for x in numbers if x % 2 != 0}

print("Odd numbers and their cubes:", cube_dict)
```

- We used dictionary comprehension to create the dictionary in a single line.
- x: x**3 means the key is x and the value is its cube.
- The condition if x % 2 != 0 ensures that only odd numbers are included.
- This produces a dictionary of odd numbers and their cubes.
